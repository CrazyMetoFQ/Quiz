```python
class Card:
    
    def __init__(self, value, color, kind='normal'):
        
        """
        value and color will make card
        """

        self.color = color.capitalize()
        self.name = f'{color.capitalize()} {value.capitalize()}'
        self.value = values_dict[value.capitalize()]
        self.kind = kind

    
    def __str__(self):
        
        return tc.colored(self.name, color=self.color.lower())
    
    def __eq__(self,other):
        
        return self.value == other.value or self.color == other.color
    
```

```python
class Deck:
    
    def __init__(self):
        
        """
        creates deck
        """
        
        self.all_cards = []
    
    def add_card(self,card):
        
        self.all_cards.append(card)
    
    def make_deck(self):
        
        #with zero and wild
        for v in value_str_tup[:]:
            for c in colors_tup:
                self.add_card(Card(**{'color':c,'value':v}))

        #without zero       
        for v in value_str_tup[1:]:
            for c in colors_tup:
                self.add_card(Card(**{'color':c,'value':v}))       
           
    
    def shuffle_cards(self):
        # Note this doesn't return anything
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        # Note we remove one card from the list of all_cards
        return self.all_cards.pop() 

    def __len__(self):
        
        return len(self.all_cards)
    
```

```python
class Player:
    
    def __init__(self,name):
        
        """
        player class
        """
        
        self.name = name
        self.all_cards = []
        
    def add_cards(self,new_cards):
        """
        adds card to players deck
        """
        
        #Cheking Types to 'Extend' or 'Append'
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
        
    def deal_one(self,indextoremove):
        """
        deals one card
        """
        return self.all_cards.pop(indextoremove)
    
    def __str__(self):
         
        return [str(i) for i in [c.kind for c in d.all_cards]]
    
    def __len__(self):
        
        return len(self.all_cards)
        
```

```python
def display_cards(player_1,player_2,last_played):
    """
    displays cards of player and last played
    """
    
    #Printing New line in start
    print('\n', end = ' ')
    
    #Player 1
    tc.cprint('Cards of Player 1:', on_color='on_red')
    
    #Player1 display
    print('\n')
    for i in  player_1.all_cards: print(f"{tc.colored('|',i.color.lower())} {i} ", end='')
    tc.cprint('|',player_1.all_cards[-1].color.lower())
    print('\n',colorama.Back.LIGHTWHITE_EX, f"Player {player_1.name} has {len(player_1)}", colorama.Back.RESET )
    
    #Middle
    print('')
    print("_"*80,'\n')
    print(f"Top Deck Card Is = {last_played[0]}")
    print("_"*80)
    print('')
    
    
    #Player 2
    print('\n')
    for i in  player_2.all_cards: print(f"{tc.colored('|',i.color.lower())} {i} ", end='')
    tc.cprint('|',player_2.all_cards[-1].color.lower())
    print('\n',colorama.Back.LIGHTWHITE_EX, f"Player {player_2.name} has {len(player_2)}", colorama.Back.RESET )
    
    #End Buffer
    print('\n')
    print('-'*100)
```

```python
def cheker(g_card,c_card):
    """
    Checks g with c if they are equal
    """

    if g_card == c_card:
        return True
    else:
        return False
    
```

```python
def cardsAvailable(player,other):
    
    """
    checks if player has any card == to middle card
    """
    
    for i,c in enumerate(player.all_cards):
        
        if c == other: 
            result = (True,i)
            return result
        
        else:
            result = False
    
    return result
```

```python
def winCheck(player):
    """
    checks if player has cards
    """
    
    if len(player) > 0:
        return True
    
    else: pass
    
```

```python
def autoTurn(player,other):
    """
    does auto turn for player
    """
    
    available = []
    for i,c in enumerate(player.all_cards):
    
        if cheker(c,other):
            available = True
            return i,c
        
        else:
            available = False
    
    return available
    
```
