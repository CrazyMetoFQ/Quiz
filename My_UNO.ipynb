```python
class Card:
    
    def __init__(self,kindCard,card_type = 'Normal',SpecialNeeded = False):
       
        """
        Kind of Card (Values Will Be Dicided Upon This)
        
        Color Value(string)
            or
        Color Type
        
        """
     
        self.kind = kindCard
        self.type = card_type
        
        #Color Check 
        def ColorAssignment(self):
            if 'Red' in kindCard.capitalize() :
                self.color = 'Red'
                self.ColorValue = 10
            if 'Green' in kindCard.capitalize()  :
                self.color = 'Green'
                self.ColorValue = 20
            if 'Blue' in kindCard.capitalize()  :
                self.color = 'Blue'
                self.ColorValue = 30
            if 'Yellow' in kindCard.capitalize()  :
                self.color = 'Yellow'
                self.ColorValue = 40
        
        #running the thing that is ^^Above^^ (to assign values)
        
        ColorAssignment(self)
        
        # Value
        def ValueAssignment(self):
            if 'Zero' in kindCard:
                self.value = 0
            if 'One' in kindCard:
                self.value = 1
            if 'Two' in kindCard:
                self.value = 2
            if 'Three' in kindCard:
                self.value = 3
            if 'Four' in kindCard:
                self.value = 4
            if 'Five' in kindCard:
                self.value = 5
            if 'Six' in kindCard:
                self.value = 6
            if 'Seven' in kindCard:
                self.value = 7
            if 'Eight' in kindCard:
                self.value = 8
            if 'Nine' in kindCard:
                self.value = 9
        
        #running the thing that is ^^Above^^ (to assign values)
        
        ValueAssignment(self)
        
        #Wether u want special card or not
        if SpecialNeeded:
            
            #Special Check

            def SpecialCheck(self):
                if card_type != 'Normal' and card_type == 'Special':

                    if 'Reverse' in kindCard  :
                        self.type = 'Reverse'
                    if 'Skip' in kindCard:
                        self.type = 'Skip'
                    if 'Add2' in kindCard.capitalize()  :
                        self.type = 'Add2'
                    if 'Wild' in kindCard :
                        self.type = 'Wild'
                    if 'Wild+4' in kindCard :
                        self.type = 'Wild+4'   

            #running the thing that is ^^Above^^ (to assign values)
            SpecialCheck(self)
    
    
    #Return The Kind of Card
    def re_kind(self):
        return self.kind   
    
    # String ( For Printing )
    def __str__(self, DoUWantColor = True ,returnornot = True):
            
        if DoUWantColor:
            #Color Printing !!! YaYaYaY

            if self.color == 'Red' :
                return Fore.RED + self.kind
            elif self.color == 'Green' :
                return Fore.GREEN + self.kind
            elif self.color == 'Blue' :
                return Fore.BLUE + self.kind
            elif self.color == 'Yellow' :
                return Fore.YELLOW + self.kind
            else:
                return 'Please Put Valid Input'
            
            if returnornot:
                return self.kind
        
        #If U Dont Want Color :-(
        else:
            return self.kind
```

```python
class Deck:
    
    """
    Makes a Simple Deck
    """
    def __init__(self,SpecialsWanted = False):
        
        """
        Creates A Deck of UNO Cards
        """
        # for Normal cards
        self.all_cards = []
        
        
        def NormalDeck_Zero(self):

             # The normal card deck Without "Zero(0)"
            for i in KindNormal0:
                self.all_cards.append(Card(i))
                
        def NormalDeck_WithoutZero(self):

            # The normal card deck Without "Zero(0)"
            for i in KindNormalNot0:
                self.all_cards.append(Card(i))
        
        if SpecialsWanted:
            #only if u wnat specials
            def SpecialDeck(self):

                # Adding Special Cards 
                for card_type in SpecialsTuple :
                    self.all_cards.append(Card(card_type,'Special'))
                    self.all_cards.append(Card(card_type,'Special'))

            def WildDeck(self):

                # Adding Special Cards 
                for card_type1 in Wilds :
                    self.all_cards.append(Card(card_type1,'Special'))
                    self.all_cards.append(Card(card_type1,'Special'))
                    self.all_cards.append(Card(card_type1,'Special'))
                    self.all_cards.append(Card(card_type1,'Special'))

        #Executing
        NormalDeck_WithoutZero(self)
        NormalDeck_Zero(self)
        
        if SpecialsWanted:
            #Just so no ERROR comes
            SpecialDeck(self)
            WildDeck(self)
                
    def shuffle(self):
        # Note this doesn't return anything
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        # Note we remove one card from the list of all_cards
        return self.all_cards.pop()        
    
    
    def deal_Multi(self,NotoRemove):  
    #Same Thing as deal_one But multiple times
        x = 0
        Deleted_Items_List = []
        while x <= NotoRemove - 1:
            x += 1
            Deleted_Item = self.all_cards.pop(0)
            Deleted_Items_List.append(Deleted_Item)
        return Deleted_Items_List
```

```python
class PlayerClass:
    
    def __init__(self,name,HumanOrComputer):
        
        """
        Class for Players
        """
        
        self.name = name
        # A new player has no cards
        self.all_cards = [] 
        
        #player or Computer
        self.player_type = HumanOrComputer
    
    
    def add_cards(self,new_cards):
        
        #Cheking Types to 'Extend' or 'Append'
        
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
            
    
    
    def deal_one(self,indextoremove):
        """
        give Index
        """
        # Note we remove one card from the list of all_cards
        return self.all_cards.pop(indextoremove)    
    
    
    def deal_Multi(self,NotoRemove):  
        """
        Does not take index just how many to deal starting from top
        """
    #Same Thing as deal_one But multiple times
        x = 0
        Deleted_Items_List = []
        while x <= NotoRemove - 1:
            x += 1
            Deleted_Item = self.all_cards.pop(0)
            Deleted_Items_List.append(Deleted_Item)
        return Deleted_Items_List
            
    
    
    def __str__(self,WhatToPrint = 'len', ColorOrNot = True, WantToPrint = False):
        
        """
        What To Print(Len or Cards)
        ColorOrNot(True/False)
        WandToPrint(True/False)
        """
        
        #Returning len
        if "LEN" in WhatToPrint.upper() :
            return f'Player {self.name} has {len(self.all_cards)} cards.'
        
        
        #For List Of Cards
        
        #If U dont Want Color
        elif "CARD" in WhatToPrint.upper() and not ColorOrNot:
            
            CARD_LIST = []
            for i in self.all_cards:
                
                if WantToPrint:
                    print(i.__str__(False))
                
                CARD_LIST.append(i.__str__(False))
            
            return CARD_LIST
        
        
        #If U Want Color
        elif "CARD" in WhatToPrint.upper() and ColorOrNot:
            
            CARD_LIST = []
            for i in self.all_cards:
                
                if WantToPrint:
                    print(i)
                CARD_LIST.append(i.__str__(True))
            
            return CARD_LIST
         
        
        else:
            print('Please Put Valid Input' 
                  '\n''\n'
                  'Total Number Of Cards The Player Has("len")'
                  '\n'  '            or         '   '\n'
                  'To Print the "Name of Cards"("Cards")')
        
        return False,"Invalid Input"
    
    #lenght
    def __len__(self):
        return len(self.all_cards)
```

```python
def PlayerNameTaker(NamesRequired = False, Names = None):
    """
    Names Should Be Given As List seperated by comma

    """
    
    #Asking Names
    
    if NamesRequired and Names == None:
        print('Player 1 is Computer''\n''\n''Player 2 is You''\n')
        Names = tuple(input('What Are The Names of Players: ').split(sep = ","))
    
    elif Names == None:
        Names = ('Computer','Player')
    
    
    #Tuple Unpacking
    (Player_Computer_Name,Player_Human_Name) = Names
    
    ComputerPlayer = PlayerClass(Player_Computer_Name,'Computer')
    HumanPlayer = PlayerClass(Player_Human_Name,'Human')
    
    return ComputerPlayer,HumanPlayer
```

```python
def AskForInput(cards_of_player_1,cards_of_player_2, Played_Cards ,Whose_To_Display = '2' 
                , InFrUs = None):
    
    """
    Asking Input From User And Confirming If it is in their hand
    
    Names Are Self Explainatory
    
    """
    
    #First Displaying Card
    displayCards(cards_of_player_1,cards_of_player_2, Played_Cards , Whose_To_Display,False)
    
    #If InFrUs(Input From User) is Not given
    if InFrUs == None:
        InFrUs = input(Fore.RESET + "Please Select Card( Name Of It Or Index( Starting From 0 )) : ")
    
    #Just So the 'If' sequence ends
    else:
        pass
    
    #To quit
    if InFrUs == '' or InFrUs == ' ' or InFrUs == 'quit' or InFrUs == 'out':
        print ('Are U Sure U Dont Want To Play')
        Conifrmation = input('True/False')
        if bool(Conifrmation):
            return "Thank U For Playing"
    
    if InFrUs == 'qq' :
        return 'toodle loo'
    
    #If Srting given 
    if InFrUs in cards_of_player_2.__str__('cards',False):
        return [True,InFrUs]
    
    #If Index Given
    elif InFrUs.isdigit or int(InFrUs) in range(len(cards_of_player_2.__str__('cards'))): 
        return [True,cards_of_player_2.__str__('cards',False)[int(InFrUs)]]
    
    else:
        print('Invalid Input')
        return False
```

```python
#Checks if two cards are equal
def checker(Card1,Card2,PrintOrNot = True):
    """
    Only For Type <Card Class>

    """
    
    if Card1.color == Card2.color or Card1.value == Card2.value:
        if PrintOrNot: 
            print('Valid')
            return True
        else:
            return True
    else:
        if PrintOrNot:
            print("Invalid")
            return False
        else:
            return False
```
