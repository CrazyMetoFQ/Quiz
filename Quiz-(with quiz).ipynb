```python
import Handy as hand

import termcolor as tc

import datetime
import time

import random as rd
```

```python
class question:
    """
    Class for Question
    """
    
    def __init__(self, q, opts, ans, g_time = 10):
        
        """
        q is question
        opt is answer 
        ans is answer
        g_time is  given time
        """
        #try effects intead of lenghty color
        
        self.q = q
        self.opts = opts
        self.ans = ans
        self.g_time = g_time
        
        #opts ref dict
        l1 = list(range(1,len(self.opts) +1))
        l2 = [str(i) for i in range(1,len(self.opts) +1)]
        l3 = [hand.Convert_AlphaNum(i,True) for i in l1]
        l4 = [i.lower() for i in self.opts]

        self.opts_ref = dict(zip(l1,(zip(l2,l3,l4))))
        
    
    def show(self):
        
        """
        show question + options 
        """
        
        print(f"U have {self.g_time} seconds")
        print(self.q,'\n')
        for x,y in zip(range(1,len(self.opts)+1),self.opts):
            tc.cprint(f'{x}: {y}', color = 'magenta')
        print(end='\n')
    
   
    
    
    def ask(self,want_time = False):
        
        """
        asks questions
        """
        def asker(self):
            
            #input from user 
            g_ans = input(tc.colored("Select Option: ",'cyan')).lower()
            self.g_ans = g_ans
        
        def ask_time_deco(self,func): 
            
            #using time delta to get time taken 
            t1 = datetime.datetime.now()
            d1 = datetime.timedelta(minutes=t1.minute, 
                                    seconds=t1.second).total_seconds()
            func(self)

            t2 = datetime.datetime.now()
            d2 = datetime.timedelta(minutes=t2.minute,
                                    seconds=t2.second).total_seconds()
            self.t_time = d2-d1
        
        if want_time: ask_time_deco(self,asker)
        else: asker(self)
    
    
    def check(self, want_time = False):
        
        """
        checks if g_ans is correct
        """
        
        def base_check(self):
            #Validating
            # Checking if its correct by comparing to opt ref ans possible values
            if self.g_ans in list(self.opts_ref.values())[self.ans-1]:
                result = True

            else:
                result = False
                
            
            return result
        
        def check_time_deco(self,func):
            
            result = func(self)
            if not result:
                tc.cprint('Incorrect','red')
                return False
            else:
                if self.g_time+1 > self.t_time:
                    tc.cprint('Correct','green')
                    return True
                else:
                    tc.cprint('Incorrect, U ran out of time','red')
                    return False
                
        if want_time: 
            self.Output = check_time_deco(self,base_check)
        else:
            self.Output = base_check(self)
    
    def exe(self):
        
        """
        executes the question completly
        """      
        #ask not taking parameters
        self.show()
        self.ask(True)
        self.check(True)
    
    def info(self):
        
        """
        infomation regarding question
        """
        #try getting which type of answer given so it can match
        
        print('The Question Was: ''\n')
        self.show()
        tc.cprint(f'\n\nYour answer was: {self.g_ans} and u took {self.t_time} seconds', 'yellow')
        tc.cprint(f' The Correct answer was: {self.ans} and given time was {self.g_time} seconds', 'green')
        #Answer correct or wrong
        if self.Output == True:
            tc.cprint(f'\nUr answer was correct',on_color='on_green')
        else:
            tc.cprint('\nUr Answer was wrong',on_color='on_red' )
    
    def __str__(self, code = False, color_want = True):
        
        """
        prints question optins and answer
        
        code = true
            question(args)
        color = True
            colored
        """
        
        if not code:
            
            qpart = f"U have {self.g_time} seconds\n{self.q}:\n\n"
            l1 = zip(range(1,len(self.opts)+1),self.opts)
            if color_want:
                l2 = [tc.colored(f'{x}: {y}', color = 'magenta') for x,y in l1]
            else:
                l2 = [f'{x}: {y}' for x,y in l1]
            optpart = hand.strlist(l2,True)
            
            anspart = f'\n\nThe Answer is {self.ans}'
            
            return qpart + optpart + anspart
        
        else:
            return f"question('{self.q}',{self.opts},{self.ans},{self.g_time})"
```

```python
class quiz:
    what = 'quiz'
    
    def __init__(self,q_ls = []):
        
        """
        q_ls --> list of questions
        """
        #For suffle 
        self.og_qs = [i for i in q_ls]
        self.qs = [i for i in q_ls]
        
    def add(self,q):
        """
        q --> ('question',[options],ans,time)
        or
        type(q) == type(question)
        """
        
        if type(q) == type(question):
            self.qs.append(q)
        else:
            self.qs.append(question(q[0],q[1],q[2],q[3]))
    
    def play(self,pause = (True,0.5),w_shuffle = True):
        
        """
        executes the entire question
        pause --> (bool,int)
            Bool,seconds
            for pause
        shuffle --> bool
            want to shuffle list 
        """
        
        if w_shuffle:
            rd.shuffle(self.qs)
        else:
            pass
        
        for i in self.qs:
            i.exe()
            print('\n---------------------------------------------\n')
            if pause[0]:
                time.sleep(pause[1])
            else:
                pass        
        
    def stats(self,pause = (True,0.5)):
        """
        stats for all question
        pause --> (bool,int)
            Bool,seconds
            for pause
        """
        
        for sn,i in zip(range(1,len(self.qs)+1),self.qs):
            
            print('\n--------------\n')
            print(f"This is Question Number: {sn}")
            i.info()
            print('\n--------------\n')
            if pause[0]:
                time.sleep(pause[1])
            else:
                pass    
            
```
